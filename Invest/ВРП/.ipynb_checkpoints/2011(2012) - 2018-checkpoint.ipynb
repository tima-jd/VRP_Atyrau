```python
import pandas as pd

import psycopg2
import io
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:2020@localhost:5432/testdb')

csv = pd.read_sql('SELECT * FROM vrp_ekz_fact ORDER BY id',engine)
csv_vrp = pd.read_sql('SELECT * FROM vrp_fact ORDER BY id',engine)


Gva_agr = csv_vrp['o_vrp_agro']
Gva_con = csv_vrp['o_vrp_building']
Gva_elc = csv_vrp['o_vrp_energo']
Gva_est = csv_vrp['o_vrp_estate']
Gva_imt = csv_vrp['o_vrp_info']
Gva_man = csv_vrp['o_vrp_man']
Gva_min = csv_vrp['o_vrp_mining']
Gva_oth = csv_vrp['o_vrp_other']
Gva_ret = csv_vrp['o_vrp_retail']
Gva_trn = csv_vrp['o_vrp_transp']
Gva_whl = csv_vrp['o_vrp_trade']
Gva_wtr = csv_vrp['o_vrp_water']

tax = csv_vrp['o_vrp_tax']

Igva_agr = csv_vrp['o_ifo_agro']
Igva_con = csv_vrp['o_ifo_building']
Igva_elc = csv_vrp['o_ifo_energo']
Igva_est = csv_vrp['o_ifo_estate']
Igva_imt = csv_vrp['o_ifo_info']
Igva_man = csv_vrp['o_ifo_man']
Igva_min = csv_vrp['o_ifo_mining']
Igva_oth = csv_vrp['o_ifo_other']
Igva_trn = csv_vrp['o_ifo_transp']
Igva_wtr = csv_vrp['o_ifo_water']
Igva_ind = csv_vrp['o_ifo_ind']
Igva_trd = csv_vrp['o_ifo_all_trade']
Igva_tot = csv_vrp['o_ifo']

Gva_ind = Gva_min+Gva_man+Gva_elc+Gva_wtr
Gva_gds = Gva_agr + Gva_ind + Gva_con
Gva_trd = Gva_ret + Gva_whl
Gva_srv = Gva_trd + Gva_trn + Gva_imt + Gva_est + Gva_oth
Grp = Gva_gds + Gva_srv

Gva_total = Grp + tax

Igva_gds = (Gva_agr.shift(1)*Igva_agr + Gva_ind.shift(1)*Igva_ind + Gva_con.shift(1)*Igva_con)/Gva_gds.shift(1)
Igva_srv = (Gva_trd.shift(1)*Igva_trd + Gva_trn.shift(1)*Igva_trn + Gva_imt.shift(1)*Igva_imt + Gva_est.shift(1)*Igva_est + Gva_oth.shift(1)*Igva_oth)/Gva_srv.shift(1)

Igrp = (Igva_gds*Gva_gds.shift(1) + Igva_srv*Gva_srv.shift(1))/Grp.shift(1)
Itax = (Igva_tot * Gva_total.shift(1) - Grp.shift(1) * Igrp)/tax.shift(1) 

Dgva_agr = Gva_agr/Gva_agr.shift(1)/Igva_agr*10000
Dgva_con = Gva_con/Gva_con.shift(1)/Igva_con*10000
Dgva_elc = Gva_elc/Gva_elc.shift(1)/Igva_elc*10000
Dgva_est = Gva_est/Gva_est.shift(1)/Igva_est*10000
Dgva_imt = Gva_imt/Gva_imt.shift(1)/Igva_imt*10000
Dgva_man = Gva_man/Gva_man.shift(1)/Igva_man*10000
Dgva_min = Gva_min/Gva_min.shift(1)/Igva_min*10000
Dgva_oth = Gva_oth/Gva_oth.shift(1)/Igva_oth*10000
Dgva_trn = Gva_trn/Gva_trn.shift(1)/Igva_trn*10000
Dgva_wtr = Gva_wtr/Gva_wtr.shift(1)/Igva_wtr*10000
Dgva_ind = Gva_ind/Gva_ind.shift(1)/Igva_ind*10000
Dgva_trd = Gva_trd/Gva_trd.shift(1)/Igva_trd*10000
Dgva_srv = Gva_srv/Gva_srv.shift(1)/Igva_srv*10000
Dgva_gds = Gva_gds/Gva_gds.shift(1)/Igva_gds*10000

Dgrp = Grp/Grp.shift(1)/Igrp*10000
Dtax = tax/tax.shift(1)/Itax*10000

Dtot = Gva_total/Gva_total.shift(1)/Igva_tot*10000

newdf = pd.DataFrame(columns=[
'Nagr_vrp',
'Ncon_vrp',
'Nelc_vrp',
'Nest_vrp',
'Nimt_vrp',
'Nman_vrp',
'Nmin_vrp',
'Noth_vrp',
'Ntrn_vrp',
'Nwtr_vrp',
'Nind_vrp',
'Ntrd_vrp',
'Nret_vrp',
'Nwhl_vrp',
'Nsrv_vrp',
'Ngds_vrp',
'Ntotal_vrp',

'lagr_vrp',
'lcon_vrp',
'lelc_vrp',
'lest_vrp',
'limt_vrp',
'lman_vrp',
'lmin_vrp',
'loth_vrp',
'ltrn_vrp',
'lwtr_vrp',
'lind_vrp',
'ltrd_vrp',
'lsrv_vrp',
'lgds_vrp',
'ltotal_vrp',

'Dagr_vrp',
'Dcon_vrp',
'Delc_vrp',
'Dest_vrp',
'Dimt_vrp',
'Dman_vrp',
'Dmin_vrp',
'Doth_vrp',
'Dtrn_vrp',
'Dwtr_vrp',
'Dind_vrp',
'Dtrd_vrp',
'Dsrv_vrp',
'Dgds_vrp',
'Dtotal_vrp',

    'Grp',
    'Tax',    
    'Igrp',
    'Itax',
    'Dgrp',
    'Dtax'
])


newdf.Nagr_vrp = Gva_agr 
newdf.Ncon_vrp = Gva_con 
newdf.Nelc_vrp = Gva_elc 
newdf.Nest_vrp = Gva_est 
newdf.Nimt_vrp = Gva_imt 
newdf.Nman_vrp = Gva_man 
newdf.Nmin_vrp = Gva_min 
newdf.Noth_vrp = Gva_oth 
newdf.Ntrn_vrp = Gva_trn 
newdf.Nwtr_vrp = Gva_wtr 
newdf.Nind_vrp = Gva_ind 
newdf.Ntrd_vrp = Gva_trd 
newdf.Nret_vrp = Gva_ret 
newdf.Nwhl_vrp = Gva_whl 
newdf.Nsrv_vrp = Gva_srv 
newdf.Ngds_vrp = Gva_gds 
newdf.Ntotal_vrp = Gva_total 


newdf.lagr_vrp = Igva_agr 
newdf.lcon_vrp = Igva_con 
newdf.lelc_vrp = Igva_elc 
newdf.lest_vrp = Igva_est 
newdf.limt_vrp = Igva_imt 
newdf.lman_vrp = Igva_man 
newdf.lmin_vrp = Igva_min 
newdf.loth_vrp = Igva_oth 
newdf.ltrn_vrp = Igva_trn 
newdf.lwtr_vrp = Igva_wtr 
newdf.lind_vrp = Igva_ind 
newdf.ltrd_vrp = Igva_trd 
newdf.lsrv_vrp = Igva_srv 
newdf.lgds_vrp = Igva_gds 
newdf.ltotal_vrp = Igva_tot 

newdf.Dagr_vrp = Dgva_agr 
newdf.Dcon_vrp = Dgva_con 
newdf.Delc_vrp = Dgva_elc 
newdf.Dest_vrp = Dgva_est 
newdf.Dimt_vrp = Dgva_imt 
newdf.Dman_vrp = Dgva_man 
newdf.Dmin_vrp = Dgva_min 
newdf.Doth_vrp = Dgva_oth 
newdf.Dtrn_vrp = Dgva_trn 
newdf.Dwtr_vrp = Dgva_wtr 
newdf.Dind_vrp = Dgva_ind 
newdf.Dtrd_vrp = Dgva_trd 
newdf.Dsrv_vrp = Dgva_srv 
newdf.Dgds_vrp = Dgva_gds 
newdf.Dtotal_vrp = Dtot 

newdf.Grp = Gva_tot
newdf.Tax = tax
newdf.Igrp = Igrp
newdf.Itax = Itax
newdf.Dgrp = Dgrp
newdf.Dtax = Dtax

newdf.to_sql('vrp_fact_ao', engine, schema='public',if_exists='replace', index = False)
```
