```python
import pandas as pd

import psycopg2
import io
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:2020@localhost:5432/testdb')

csv = pd.read_sql('SELECT * FROM vrp_prognoz_kzylkoga ORDER BY index',engine)


Nagr = csv['Nagr_kzylkoga']
Ncon = csv['Ncon_kzylkoga']
Nind = csv['Nind_kzylkoga']

n = Nind + Ncon + Nagr
Ngds = pd.DataFrame(columns=['Ngds_kzylkoga'])

Ngds.Ngds_kzylkoga = n


Ngds.to_sql('ngds_prog_kzylkoga', engine, schema='public',if_exists='replace', index = True)
```
