```python
import pandas as pd

import psycopg2
import io
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:2020@localhost:5432/testdb')

csv = pd.read_sql('SELECT * FROM vrp_ekz_fact ORDER BY id',engine)

lwtr = csv['o_ifo_water_kzylkoga']
lelc = csv['o_ifo_energo_kzylkoga']
lman = csv['o_ifo_man_kzylkoga']
lmin = csv['o_ifo_mining_kzylkoga']

Nwtr = csv['o_prod_water_kzylkoga']
Nelc = csv['o_prod_energo_kzylkoga']
Nman = csv['o_prod_man_kzylkoga']
Nmin = csv['o_prod_mining_kzylkoga']

Nind = Nmin+Nman+Nelc+Nwtr

newdf = pd.DataFrame(columns=['Dind_kzylkoga','lind_kzylkoga','Nind_kzylkoga'], index=range(19))
i=1
while i<=18:    
    
    lind = (lmin[i]*Nmin[i-1]+lman[i]*Nman[i-1]+lelc[i]*Nelc[i-1]+lwtr[i]*Nwtr[i-1])/Nind[i-1]    
    Dind=(Nind[i]/Nind[i-1]/lind)*10000
    
    newdf.loc[i].Dind_kzylkoga = Dind
    newdf.loc[i].lind_kzylkoga = lind
    i= i+1
    
lind = newdf['lind_kzylkoga']
Dind = newdf['Dind_kzylkoga']
newdf.Nind_kzylkoga = Nind
Nind = newdf['Nind_kzylkoga']

newdf.to_sql('dind_lind_nind_fact_kzylkoga', engine, schema='public',if_exists='replace', index = True)
```
