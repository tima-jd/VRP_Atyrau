```python
import pandas as pd

csv = pd.read_csv('vrp_ekz_fact.csv')
csv_p = pd.read_csv('vrp_ekz_prog.csv')

lret_csv = pd.read_csv('lret(2019-2025).csv')
lwhl_csv = pd.read_csv('lwhl(2019-2025).csv')
lagr_csv = pd.read_csv('lagr(2019-2025).csv')
lmin_csv = pd.read_csv('lmin(2019-2025).csv')
lcon_csv = pd.read_csv('lcon(2019-2025).csv')
ltrd_csv = pd.read_csv('ltrd(2001-2018).csv')
man_csv = pd.read_csv('lman, Nman(2019-2025).csv')
Ncon_csv = pd.read_csv('Ncon(2019-2025).csv')
Ntrd_csv = pd.read_csv('Ntrd(2001-2018).csv')
Nmin_csv = pd.read_csv('Nmin(2019-2025).csv')
Nret_csv = pd.read_csv('Nret(2019-2025).csv')
Nagr_csv = pd.read_csv('Nagr(2019-2025).csv')

Nret_f = csv['o_prod_retail_atyrau']
Nwhl_f = csv['o_prod_trade_atyrau']
Nagr_f = csv['o_prod_agro_atyrau']
Nmin_f = csv['o_prod_mining_atyrau']
Nman_f = csv['o_prod_man_atyrau']
Ncon_f = csv['o_prod_building_atyrau']
Nelc_f = csv['o_prod_energo_atyrau']
Nwtr_f = csv['o_prod_water_atyrau']
Nind_f = csv['o_prod_water_atyrau']

lelc_f = csv['o_ifo_energo_atyrau']
lwtr_f = csv['o_ifo_water_atyrau']


Ntrd_f = Ntrd_csv['0']
ltrd_f = ltrd_csv['0']

Nmin = Nmin_csv['Nmin_b']
Nmin = Nmin_csv['Nmin_o']
Nmin = Nmin_csv['Nmin_p']


Nman_b =  man_csv['Nman_b']
Nman_o =  man_csv['Nman_o']
Nman_p =  man_csv['Nman_p']

lman_b =  man_csv['lman_b']
lman_o =  man_csv['lman_o']
lman_p =  man_csv['lman_p']

lret_b = lret_csv['lret_b']
lret_o = lret_csv['lret_o']
lret_p = lret_csv['lret_p']

lwhl_b = lwhl_csv['lwhl_b']
lwhl_o = lwhl_csv['lwhl_o']
lwhl_p = lwhl_csv['lwhl_p']

lagr_b = lagr_csv['lagr_b']
lagr_o = lagr_csv['lagr_o']
lagr_p = lagr_csv['lagr_p']

lmin_b = lmin_csv['lmin_b']
lmin_o = lmin_csv['lmin_o']
lmin_p = lmin_csv['lmin_p']

lcon_b = lcon_csv['lcon_b']
lcon_o = lcon_csv['lcon_o']
lcon_p = lcon_csv['lcon_p']

Ncon_b = Ncon_csv['Ncon_b']
Ncon_o = Ncon_csv['Ncon_o']
Ncon_p = Ncon_csv['Ncon_p']

Nret_b = Nret_csv['Nret_b']
Nret_o = Nret_csv['Nret_o']
Nret_p = Nret_csv['Nret_p']

Nagr_b = Nagr_csv['Nagr_b']
Nagr_o = Nagr_csv['Nagr_o']
Nagr_p = Nagr_csv['Nagr_p']

inf = csv_p['o_inflation']

ltrd_b = (lret_b[0]*Nret_f[18] + lwhl_b[0]*Nwhl_f[18])/Ntrd_f[18]
lammct_b = (lagr_b[0]*Nagr_f[18] + lmin_b[0]*Nmin_f[18] + lman_b[0]*Nman_f[18] + lcon_b[0]*Ncon_f[18] + ltrd_b * Ntrd_f[18])/(Nagr_f[18]+Nmin_f[18]+Nman_f[18]+Ncon_f[18]+Ntrd_f[18])
S_b= (100*((lammct_b-100)*1.5024317564210523 +100))/100
K_b= (100*((lammct_b-100)*0.6356716742567134 +100))/100
lelc_b= S_b/100 * 100
lwtr_b= K_b/100 * 100
Delc_b = inf * 0.9524230169082659 + 100
Dwtr_b = inf * 1.0248473801986724 + 100
Nelc_b = (Nelc_f[18]*lelc_b*Delc_b[0])/10000
Nwtr_b = (Nwtr_f[18]*lwtr_b*Dwtr_b[0])/10000
Nind_b = Nmin_b[0]+Nman_b[0]+Nelc_b+Nwtr_b
Nind_f = Nmin_f+Nman_f+Nelc_f+Nwtr_f
lind_b = ((lmin_b[0]*Nmin_f[18])+(lman_b[0]*Nman_f[18])+(lelc_b*Nelc_f[18])+(lwtr_b*Nwtr_f[18]))/ Nind_f[18]
Nind_b = Nmin_b[0]+Nman_b[0]+Nelc_b+Nwtr_b
Dind_b= (Nind_b/Nind_f[18]/lind_b)*10000
Dwhl_b = inf[0] * 0.9627018135086216 + (Dind_b-100)*1.0043958806337958+100
Nwhl_b = (Nwhl_f[18]*lwhl_b[0]*Dwhl_b)/10000
Ntrd_b = Nret_b[0]+Nwhl_b
Dtrd_b = Ntrd_b/(Nret_f[18] + Nwhl_f[18])/ltrd_b *10000
```

```python
i = 0

newdf = pd.DataFrame(columns=['ltrd','lelc','lwtr','lind','Ntrd','Nelc','Nwtr','Nind','Nwhl','Dtrd','Delc','Dwtr','Dwhl','Dind'],index = range(21))

while i<=19:
    Nind_p = Nmin[i]+Nman[i]+Nelc+Nwtr
    ltrd = (lret[i+1]*Nret[i] + lwhl[i+1]*Nwhl)/Ntrd
    lammct = (lagr[i+1]*Nagr[i] + lmin[i+1]*Nmin[i] + lman[i+1]*Nman[i] + lcon[i+1]*Ncon[i] + ltrd * Ntrd)/(Nagr[i]+Nmin[i]+Nman[i]+Ncon[i]+Ntrd)
    
    S= (100*((lammct-100)*1.5024317564210523 +100))/100
    K = (100*((lammct-100)*0.6356716742567134 +100))/100
    
    lelc= S/100 * 100
    lwtr= K/100 * 100
    
    Nelc = (Nelc*lelc*Delc[i+1])/10000
    Nwtr = (Nwtr*lwtr*Dwtr[i+1])/10000
    Nind = Nmin[i+1]+Nman[i+1]+Nelc+Nwtr
    lind = ((lmin[i+1]*Nmin[i])+(lman[i+1]*Nman[0])+(lelc*Nelc)+(lwtr*Nwtr))/ Nind
    Nind = Nmin[i+1]+Nman[i+1]+Nelc+Nwtr    
    Dind= (Nind/Nind_p/lind)*10000
    Dwhl = inf[i+1] * 0.9627018135086216 + (Dind-100)*1.0043958806337958+100
    
    Ntrd_l = Nret[i]+Nwhl
    Nwhl = (Nwhl*lwhl[i+1]*Dwhl)/10000
    Ntrd = Nret[i+1]+Nwhl   
    
    Dtrd = Ntrd/Ntrd_l/ltrd*10000
    
    i = i+1
    newdf.loc[i].ltrd = ltrd
    newdf.loc[i].lelc = lelc
    newdf.loc[i].lwtr = lwtr
    newdf.loc[i].lind = lind
    newdf.loc[i].Ntrd = Ntrd
    newdf.loc[i].Nelc = Nelc
    newdf.loc[i].Nwtr = Nwtr
    newdf.loc[i].Nind = Nind
    newdf.loc[i].Nwhl = Nwhl
    
    newdf.loc[i].Dtrd = Dtrd     
    newdf.loc[i].Dwhl = Dwhl
    newdf.loc[i].Dind = Dind
    
newdf.loc[0].ltrd = 103.65167233494977
newdf.loc[0].lelc = 105.5774617579105
newdf.loc[0].lwtr = 102.35979733428915
newdf.loc[0].lind = 103.18824883806036

newdf.loc[0].Ntrd = 3515624.776256665
newdf.loc[0].Nelc = 49847.5824624706
newdf.loc[0].Nwtr = 9807.108915579907
newdf.loc[0].Nind = 627049.6310521336
newdf.loc[0].Nwhl = 3245684.6068475866

newdf.loc[0].Dtrd = 105.21241187086572
newdf.loc[0].Dwhl = 105.06484539102014
newdf.loc[0].Dind = 99.29174790165139
    
newdf.Delc = Delc 
newdf.Dwtr = Dwtr

lind = newdf['lind']
lelc = newdf['lelc']
lwtr = newdf['lwtr']
ltrd = newdf['ltrd']

Nind = newdf['Nind']
Nelc = newdf['Nelc']
Nwtr = newdf['Nwtr']
Ntrd = newdf['Ntrd']
Nwhl = newdf['Nwhl']
```
