```python
import pandas as pd

import psycopg2
import io
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:2020@localhost:5432/testdb')

csv = pd.read_sql('SELECT * FROM vrp_prognoz_atyrau ORDER BY index',engine)


Nagr = csv['Nagr_atyrau']
Ncon = csv['Ncon_atyrau']
Nind = csv['Nind_atyrau']

n = Nind + Ncon + Nagr
Ngds = pd.DataFrame(columns=['Ngds_atyrau'])

Ngds.Ngds_atyrau = n

Ngds.to_sql('ngds_prog_atyrau', engine, schema='public',if_exists='replace', index = True)
```
