```python
import pandas as pd

import psycopg2
import io
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:2020@localhost:5432/testdb')

csv = pd.read_sql('SELECT * FROM vrp_prognoz_kurmangazy ORDER BY index',engine)


Nagr = csv['Nagr_kurmangazy']
Ncon = csv['Ncon_kurmangazy']
Nind = csv['Nind_kurmangazy']

n = Nind + Ncon + Nagr
Ngds = pd.DataFrame(columns=['Ngds_kurmangazy'])

Ngds.Ngds_kurmangazy = n

Ngds.to_sql('ngds_prog_kurmangazy', engine, schema='public',if_exists='replace', index = True)
```
