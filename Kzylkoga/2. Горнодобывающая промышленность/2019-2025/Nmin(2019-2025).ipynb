```python
import pandas as pd
import numpy as np
csv_file = pd.read_csv('vrp_ekz_prog.csv')
Dmin = pd.read_csv('Dmin(2019-2025).csv')
lmin = pd.read_csv('lmin(2019-2025).csv')

Dmin_b = Dmin['Dmin_b']
Dmin_o = Dmin['Dmin_o']
Dmin_p = Dmin['Dmin_p']

lmin_b = lmin['lmin_b']
lmin_o = lmin['lmin_o']
lmin_p = lmin['lmin_p']


newdf = pd.DataFrame(columns=['Nmin_b','Nmin_o','Nmin_p'], index=range(12))

b = 52144.0
o = 52144.0
p = 52144.0
i=0

while i<=len(lmin_b)-1:
    b = b * lmin_b[i] * Dmin_b[i]/10000
    o = o * lmin_o[i] * Dmin_o[i]/10000
    p = p * lmin_p[i] * Dmin_p[i]/10000
    
    newdf.loc[i].Nmin_b = b    
    newdf.loc[i].Nmin_o = o    
    newdf.loc[i].Nmin_p = p    
    i=i+1
newdf
```

```python
newdf.to_csv(r'Nmin_Kzylkoga(2019-2025).csv')
```
