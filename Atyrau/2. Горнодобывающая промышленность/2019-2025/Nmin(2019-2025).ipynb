```python
import pandas as pd
import numpy as np

import psycopg2
import io
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:2020@localhost:5432/testdb')
csv_file = pd.read_sql("SELECT * FROM vrp_ekz_prog",engine)

Dmin = pd.read_sql("SELECT * FROM dmin_prog_atyrau",engine)
lmin = pd.read_sql("SELECT * FROM lmin_prog_atyrau",engine)


Dmin_b = Dmin['Dmin_b']
Dmin_o = Dmin['Dmin_o']
Dmin_p = Dmin['Dmin_p']

lmin_b = lmin['lmin_b']
lmin_o = lmin['lmin_o']
lmin_p = lmin['lmin_p']


newdf = pd.DataFrame(columns=['Nmin_b','Nmin_o','Nmin_p'], index=range(12))

b = 295615
o = 295615
p = 295615
i=0

while i<=len(lmin_b)-1:
    b = b * lmin_b[i] * Dmin_b[i]/10000
    o = o * lmin_o[i] * Dmin_o[i]/10000
    p = p * lmin_p[i] * Dmin_p[i]/10000
    
    newdf.loc[i].Nmin_b = b    
    newdf.loc[i].Nmin_o = o    
    newdf.loc[i].Nmin_p = p    
    i=i+1

newdf.to_sql('nmin_prog_atyrau', engine, schema='public',if_exists='replace', index = True)

newdf
```
