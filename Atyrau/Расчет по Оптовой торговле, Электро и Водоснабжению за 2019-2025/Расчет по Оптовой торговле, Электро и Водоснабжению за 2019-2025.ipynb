```python
import pandas as pd

csv = pd.read_csv('vrp_ekz_fact.csv')
csv_p = pd.read_csv('vrp_ekz_prog.csv')

lret_csv = pd.read_csv('lret(2019-2025).csv')
lwhl_csv = pd.read_csv('lwhl(2019-2025).csv')
lagr_csv = pd.read_csv('lagrt(2019-2025).csv')
lmin_csv = pd.read_csv('lmin(2019-2025).csv')
lcon_csv = pd.read_csv('lcon(2019-2025).csv')
ltrd_csv = pd.read_csv('ltrd(2001-2018).csv')
man_csv = pd.read_csv('lman, Nman(2019-2025).csv')
Ncon_csv = pd.read_csv('Ncon(2019-2025).csv')
Ntrd_csv = pd.read_csv('Ntrd(2001-2018).csv')
Nmin_csv = pd.read_csv('Nmin(2019-2025).csv')
Nret_csv = pd.read_csv('Nret(2019-2025).csv')
Nagr_csv = pd.read_csv('Nagr(2019-2025).csv')

Nret_f = csv['o_prod_retail_atyrau']
Nwhl_f = csv['o_prod_trade_atyrau']
Nagr_f = csv['o_prod_agro_atyrau']
Nmin_f = csv['o_prod_mining_atyrau']
Nman_f = csv['o_prod_man_atyrau']
Ncon_f = csv['o_prod_building_atyrau']
Nelc_f = csv['o_prod_energo_atyrau']
Nwtr_f = csv['o_prod_water_atyrau']
Nind_f = csv['o_prod_water_atyrau']

lelc_f = csv['o_ifo_energo_atyrau']
lwtr_f = csv['o_ifo_water_atyrau']


Ntrd_f = Ntrd_csv['0']
ltrd_f = ltrd_csv['0']


Nmin = Nmin_csv['c1']

Nman = man_csv['Nman']
lman = man_csv['lman']

lret = lret_csv['c1']
lwhl = lwhl_csv['c1']
lagr = lagr_csv['c1']
lmin = lmin_csv['c1']
lcon = lcon_csv['c1']
Ncon = Ncon_csv['c1']
Nret = Nret_csv['c1']
Nagr = Nagr_csv['c1']

inf = csv_p['o_inflation']

ltrd = (lret[0]*Nret_f[18] + lwhl[0]*Nwhl_f[18])/Ntrd_f[18]

lammct = (lagr[0]*Nagr_f[18] + lmin[0]*Nmin_f[18] + lman[0]*Nman_f[18] + lcon[0]*Ncon_f[18] + ltrd * Ntrd_f[18])/(Nagr_f[18]+Nmin_f[18]+Nman_f[18]+Ncon_f[18]+Ntrd_f[18])

S= (100*((lammct-100)*1.5024317564210523 +100))/100
K = (100*((lammct-100)*0.6356716742567134 +100))/100

lelc= S/100 * 100

lwtr= K/100 * 100

Delc = inf * 0.9524230169082659 + 100
Dwtr = inf * 1.0248473801986724 + 100

Nelc = (Nelc_f[18]*lelc*Delc[0])/10000
Nwtr = (Nwtr_f[18]*lwtr*Dwtr[0])/10000

Nind = Nmin[0]+Nman[0]+Nelc+Nwtr
Nind_f = Nmin_f+Nman_f+Nelc_f+Nwtr_f
lind = ((lmin[0]*Nmin_f[18])+(lman[0]*Nman_f[18])+(lelc*Nelc_f[18])+(lwtr*Nwtr_f[18]))/ Nind_f[18]

Nind = Nmin[0]+Nman[0]+Nelc+Nwtr
Dind= (Nind/Nind_f[18]/lind)*10000

Dwhl = inf[0] * 0.9627018135086216 + (Dind-100)*0.8320900666908758+100

Nwhl = (Nwhl_f[18]*lwhl[0]*Dwhl)/10000
Ntrd = Nret[0]+Nwhl

Dtrd = Ntrd/(Nret_f[18] + Nwhl_f[18])/ltrd_f[18] *10000

Dind
```

```python
i = 0

newdf = pd.DataFrame(columns=['ltrd','lelc','lwtr','lind','Ntrd','Nelc','Nwtr','Nind','Nwhl','Dtrd','Delc','Dwtr','Dwhl','Dind'],index = range(21))

while i<=19:
    Nind_p = Nmin[i]+Nman[i]+Nelc+Nwtr
    ltrd = (lret[i+1]*Nret[i] + lwhl[i+1]*Nwhl)/Ntrd
    lammct = (lagr[i+1]*Nagr[i] + lmin[i+1]*Nmin[i] + lman[i+1]*Nman[i] + lcon[i+1]*Ncon[i] + ltrd * Ntrd)/(Nagr[i]+Nmin[i]+Nman[i]+Ncon[i]+Ntrd)
    
    S= (100*((lammct-100)*1.5024317564210523 +100))/100
    K = (100*((lammct-100)*0.6356716742567134 +100))/100
    
    lelc= S/100 * 100
    lwtr= K/100 * 100
    
    Nelc = (Nelc*lelc*Delc[i+1])/10000
    Nwtr = (Nwtr*lwtr*Dwtr[i+1])/10000
    Nind = Nmin[i+1]+Nman[i+1]+Nelc+Nwtr
    lind = ((lmin[i+1]*Nmin[i])+(lman[i+1]*Nman[0])+(lelc*Nelc)+(lwtr*Nwtr))/ Nind
    Nind = Nmin[i+1]+Nman[i+1]+Nelc+Nwtr    
    Dind= (Nind/Nind_p/lind)*10000
    Dwhl = inf[i+1] * 0.9627018135086216 + (Dind-100)*0.8320900666908758+100
    
    Ntrd_l = Nret[i]+Nwhl
    Nwhl = (Nwhl*lwhl[i+1]*Dwhl)/10000
    Ntrd = Nret[i+1]+Nwhl   
    
    Dtrd = Ntrd/Ntrd_l/ltrd*10000
    
    i = i+1
    newdf.loc[i].ltrd = ltrd
    newdf.loc[i].lelc = lelc
    newdf.loc[i].lwtr = lwtr
    newdf.loc[i].lind = lind
    newdf.loc[i].Ntrd = Ntrd
    newdf.loc[i].Nelc = Nelc
    newdf.loc[i].Nwtr = Nwtr
    newdf.loc[i].Nind = Nind
    newdf.loc[i].Nwhl = Nwhl
    
    newdf.loc[i].Dtrd = Dtrd     
    newdf.loc[i].Dwhl = Dwhl
    newdf.loc[i].Dind = Dind
    
newdf.loc[0].ltrd = 103.65167233494977
newdf.loc[0].lelc = 105.5774617579105
newdf.loc[0].lwtr = 102.35979733428915
newdf.loc[0].lind = 103.18824883806036

newdf.loc[0].Ntrd = 3520853.5296533024
newdf.loc[0].Nelc = 49847.5824624706
newdf.loc[0].Nwtr = 9807.108915579907
newdf.loc[0].Nind = 627408.0272988597
newdf.loc[0].Nwhl = 3250913.360244224

newdf.loc[0].Dtrd = 61.419537007385266
newdf.loc[0].Dwhl = 105.23410341629052
newdf.loc[0].Dind = 99.34849905501564
    
newdf.Delc = Delc 
newdf.Dwtr = Dwtr

newdf.to_csv(r'Расчет по Оптовой торговле, Электро и Водоснабжению.csv')
```
